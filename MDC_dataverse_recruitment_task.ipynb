## Sales performance analysis

### Subtask:
Calculate key sales metrics (total sales, average sale per order, etc.). Analyze sales performance over time (e.g., monthly or yearly trends). Visualize sales trends using interactive plots (Plotly).


**Reasoning**:
Calculate total sales and average sale per order, extract year and month from 'Order Date', group by year and month to calculate monthly sales, and group by year to calculate yearly total and average sales.



```python
total_sales = df['Sales'].sum()
average_sale_per_order = total_sales / df['Order ID'].nunique()
print(f"Total Sales: {total_sales:.2f}")
print(f"Average Sale per Order: {average_sale_per_order:.2f}")

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
yearly_sales = df.groupby('Order Year')['Sales'].agg(['sum', 'mean']).reset_index()
yearly_sales.rename(columns={'sum': 'Total Sales', 'mean': 'Average Sales per Order'}, inplace=True)

display(monthly_sales.head())
display(yearly_sales.head())
```

**Reasoning**:
Create an interactive line plot using Plotly to visualize the monthly sales trends over time and interactive bar plots to visualize the yearly total sales and average sales per order.



```python
import plotly.express as px

monthly_sales['Order Date'] = pd.to_datetime(monthly_sales['Order Year'].astype(str) + '-' + monthly_sales['Order Month'].astype(str))

fig_monthly = px.line(monthly_sales, x='Order Date', y='Sales', title='Monthly Sales Trends')
fig_monthly.update_layout(xaxis_title='Date', yaxis_title='Total Sales')
fig_monthly.show()

fig_yearly_total = px.bar(yearly_sales, x='Order Year', y='Total Sales', title='Yearly Total Sales')
fig_yearly_total.update_layout(xaxis_title='Year', yaxis_title='Total Sales')
fig_yearly_total.show()

fig_yearly_avg = px.bar(yearly_sales, x='Order Year', y='Average Sales per Order', title='Yearly Average Sales per Order')
fig_yearly_avg.update_layout(xaxis_title='Year', yaxis_title='Average Sales per Order')
fig_yearly_avg.show()
```
